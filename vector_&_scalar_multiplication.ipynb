```python
#Vector and scalar multiplication
import matplotlib.pyplot as pyplot
import numpy as np

vec = np.array([ 3,4,5,2])
s = 2


vec*s
```

```python
vec2d= np.array([1,2])
s1 = 2
s2 = 5
s3 =-1
```

```python
pyplot.plot([0,vec2d[0]],[0,vec2d[1]],'bs--')
pyplot.plot ([0,s1*vec2d[0]],[0,s1*vec2d[1]],'ro-',label='v*s1')
pyplot.plot ([0,s2*vec2d[0]],[0,s2*vec2d[1]],'kp-',label='v*s2')
pyplot.plot ([0,s3*vec2d[0]],[0,s3*vec2d[1]],'g*-',label='v*s3')

pyplot.axis('square')
pyplot.xlim([-4,4])
pyplot.ylim([-4,4])
pyplot.grid()
pyplot.legend()
pyplot.show()

#x = 0,vec2d[0] y= 0,vec2d[1]
```

```python
vec2d1= np.array([4,2])
vec2d2= np.array([6,3])
result_add = vec2d1+vec2d2
result_minus = vec2d1-vec2d2
print(result_add)
print(result_minus)
```

```python
pyplot.plot([0,vec2d1[0]],[0,vec2d1[1]],"bs-",label="v1")
pyplot.plot([0,vec2d2[0]],[0,vec2d2[1]],"ro-",label="v2")

pyplot.axis('square')
pyplot.xlim([-2,7])
pyplot.ylim([-2,6])
pyplot.grid()
pyplot.legend()
pyplot.show()
```

```python
pyplot.plot([0,result_add[0]],[0,result_add[1]],"bs-",label="v1+v2")
pyplot.plot([0,result_minus[0]],[0,result_minus[1]],"ro-",label="v1-v2")

pyplot.axis('square')
pyplot.xlim([-3,12])
pyplot.ylim([-3,12])
pyplot.grid()
pyplot.legend()
pyplot.show()
```
