```python
#Vector and scalar multiplication
import matplotlib.pyplot as pyplot
import numpy as np

vec = np.array([ 3,4,5,2])
s = 2


vec*s
```

```python
vec2d= np.array([1,2])
s1 = 2
s2 = 5
s3 =-1
```

```python
pyplot.plot([0,vec2d[0]],[0,vec2d[1]],'bs--')
pyplot.plot ([0,s1*vec2d[0]],[0,s1*vec2d[1]],'ro-',label='v*s1')
pyplot.plot ([0,s2*vec2d[0]],[0,s2*vec2d[1]],'kp-',label='v*s2')
pyplot.plot ([0,s3*vec2d[0]],[0,s3*vec2d[1]],'g*-',label='v*s3')

pyplot.axis('square')
pyplot.xlim([-4,4])
pyplot.ylim([-4,4])
pyplot.grid()
pyplot.legend()
pyplot.show()

#x = 0,vec2d[0] y= 0,vec2d[1]
```
